```python
# Import library yang diperlukan
import pandas as pd               # manipulasi data
import numpy as np                # operasi numerik
import matplotlib.pyplot as plt   # visualisasi
import seaborn as sns             # visualisasi tambahan
```

```python
# ============================
# 1. Memuat Dataset Titanic
# ============================
data_titanic = sns.load_dataset("titanic")

# Menampilkan 5 baris pertama dataset
print("5 Baris Pertama Dataset Titanic:")
print(data_titanic.head())
```

```python
# ============================
# 2. Modifikasi Kolom
# ============================
# Buat kolom baru 'gender_num' (male=1, female=0)
data_titanic['gender_num'] = data_titanic['sex'].map({'male': 1, 'female': 0})
print("\nContoh kolom gender_num:")
print(data_titanic[['sex', 'gender_num']].head())
```

```python
# ============================
# 3. Statistik Deskriptif
# ============================
print("\nInformasi Dataset:")
print(data_titanic.info())

print("\nStatistik Deskriptif:")
print(data_titanic.describe())
```

```python
# ============================
# 4. Analisis Missing Values
# ============================
missing_counts = data_titanic.isnull().sum()
print("\nJumlah Missing Values per Kolom:")
print(missing_counts)

# Visualisasi missing values
plt.figure(figsize=(10,5))
sns.heatmap(data_titanic.isnull(), cbar=False, cmap="viridis")
plt.title("Visualisasi Missing Values")
plt.show()
```

```python
# ============================
# 5. Visualisasi Data
# ============================

# a) Distribusi Umur Penumpang
plt.figure(figsize=(8,5))
sns.histplot(data_titanic['age'], bins=30, kde=True, color="skyblue")
plt.title("Distribusi Umur Penumpang")
plt.xlabel("Umur")
plt.ylabel("Jumlah Penumpang")
plt.show()

# b) Perbandingan Tarif (fare) berdasarkan Kelas
plt.figure(figsize=(8,5))
sns.boxplot(x="class", y="fare", data=data_titanic, palette="Set2")
plt.title("Perbandingan Tarif berdasarkan Kelas")
plt.xlabel("Kelas")
plt.ylabel("Tarif")
plt.show()

# c) Tingkat Kelangsungan Hidup berdasarkan Jenis Kelamin
plt.figure(figsize=(8,5))
sns.barplot(x="sex", y="survived", data=data_titanic, palette="pastel")
plt.title("Tingkat Kelangsungan Hidup berdasarkan Jenis Kelamin")
plt.xlabel("Jenis Kelamin")
plt.ylabel("Proporsi Selamat")
plt.show()

# d) Hubungan Umur dan Tarif dengan Warna berdasarkan Survival
plt.figure(figsize=(8,5))
sns.scatterplot(x="age", y="fare", hue="survived", data=data_titanic, palette="coolwarm", alpha=0.7)
plt.title("Hubungan Umur dan Tarif terhadap Kelangsungan Hidup")
plt.xlabel("Umur")
plt.ylabel("Tarif")
plt.show()
```
